```python
from tkinter import Tk
from tkinter import messagebox
import tkinter
import time as t
import cv2
import math
import threading
# alert message
def break_time_alert():
    sec = 0
    while True:
        sec += 10
        t.sleep(sec)
        root = Tk()
        root.withdraw()
        tkinter.messagebox.showwarning("주의","사용한 지 %d 분 지났습니다. 10분 정도 휴식을 취하세요." 
                                    % (sec/ 60))
def face_detection():
    count = 1
    distance = 0.0
    font = cv2.FONT_HERSHEY_SIMPLEX
    
    # Capture from the default deivce
    cap = cv2.VideoCapture(0)
    
    # The samples thingy
    # Make sure the file is in the same folder as the program..
    # otherwise this will not work
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

        # brightness control from distance
    def change_brightness(distance):
        if distance <= 30 and distance >= 0 :
             lights = 1
            return lights
         elif distance > 30 and distance <= 50 :
            lights = 2
            return lights
        elif distance > 50 and distance <= 70 :
            lights = 3
            return lights
        elif distance > 70 and distance <= 90 :
            lights = 4
            return lights
        elif distance > 90:
            lights = 5
            return lights
    # sleep for 5 minute after image capture
    def count_realtime(count):
        if count % 50 == 0:
            t.sleep(10)
    #     t1 = t.time()
    while True:
        #     t2 = t.time()
        #     alert_message(t2,t1)
        count_realtime(count)
        # Caputure a single frame
        ret, huge_frame = cap.read()
        frame = cv2.resize(huge_frame, (0,0), fx=1.0, fy=1.0, interpolation=cv2.INTER_NEAREST)
        # Create the greyscale and detect faces
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        # Add squeres for each face
        for (x, y, w, h) in faces:
            distancei = (2*3.14 * 180)/(w+h*360)*1000 + 3
        #        distance = distancei *2.54
            distance = math.floor(distancei*2.54)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 3)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = frame[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            for (ex,ey,ew,eh) in eyes:
                cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)

        # Display the resulting frame
            cv2.putText(frame,'D = ' + str(distance) + 'cm', (x+w,y+h),font,1,(0,255,0),2)
            cv2.putText(frame,'brights = ' + str(change_brightness(distance)),(x+w,y+h+30),font,1,(0,0,255),2)

        cv2.imshow('face detection', frame)
        count += 1
        print("D:  %d lights: %d " %(distance,change_brightness(distance)))
        if cv2.waitKey(1) == ord('q'):
            break

    # Stop the captureq
    cap.release()
    # Destory the wqqindow
    cv2.destroyAllWindows()
t1 = threading.Thread(target=break_time_alert)  
t2 = threading.Thread(target=face_detection)
t1.start()
t2.start()
```
