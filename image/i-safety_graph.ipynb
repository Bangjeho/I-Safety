```python
from matplotlib import pyplot as plt

x = [15,23,31,42,54,67,72,81,94,99]
y = [14,23,32,43,57,70,75,85,99,104]

plt.scatter(x,x)
plt.scatter(x,y)
plt.xlabel("Distance")
plt.ylabel("Distance")
plt.title("how much error?")
plt.plot(x,x, label = "real distance")
plt.plot(x,y, label = "face_detection distance")

plt.legend(loc = 9)
# plt.show()
plt.savefig("how much error my face_detection app.png")
```
